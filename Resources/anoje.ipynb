```python
ecomm_transactions_df['shipping_address'] = ecomm_transactions_df['shipping_address'].str.replace('\n', ', ')
ecomm_transactions_df['billing_address'] = ecomm_transactions_df['billing_address'].str.replace('\n', ', ')

print(ecomm_transactions_df[['shipping_address', 'billing_address']].head())
ecomm_transactions_df
```

```python
import numpy as np


mean = np.mean(ecomm_transactions_df['transaction_amount'])
standard_dev = np.std(ecomm_transactions_df['transaction_amount'])

limit = 3
upper_bound = mean + limit * standard_dev
lower_bound = mean - limit * standard_dev
 
new_frame = ecomm_transactions_df[(ecomm_transactions_df['transaction_amount'] >= lower_bound) & 
                                  (ecomm_transactions_df['transaction_amount'] <= upper_bound)]


#print(new_frame.head())
new_frame

#print(upper_bound)
#print(lower_bound)
#print(mean)


```
